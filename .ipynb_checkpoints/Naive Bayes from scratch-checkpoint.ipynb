```python
#Gender distribution
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
plt.figure(figsize=(4,4))
print(data['Gender'].value_counts().sort_index())
data['Gender'].value_counts().sort_index().plot(kind='bar', color = {'royalblue','orange'})
plt.title('Gender Distribution')
plt.xlabel('Gender');
plt.ylabel('COUNTS');
```

```python
#Bi-variate analysis Gender vs Purchased
sns.set(style="whitegrid", color_codes=True)
plt.figure(figsize=(4,4))
ax=sns.violinplot(data=data, x= 'Gender', y='Purchased')
```

```python
#correlation visualization
sns.pairplot(data, diag_kind='kde')
```

```python
#seperating dependant and independant features
x=data.iloc[:,[1,2,3]]
y=data.iloc[:,-1]
print(x.head())
print('Independant feature Purchased')
print(y.head())
```

Above plot shows a lot of randomness and no perticular correlation amongst features

```python
#Converting catagorical variable in numericals using StandardScaler
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x['Gender']=le.fit_transform(x['Gender'])
x.head()
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#normalization using StandardScaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
#training
from sklearn.naive_bayes import GaussianNB
clf=GaussianNB()
clf.fit(X_train,y_train)
```

```python
#predict
y_pred=clf.predict(X_test)
```

```python
print(y_pred)
```

```python
#bi-variate analysis y_pred vs y_test
plt.scatter(y_pred, y_test, color = "red")
plt.plot(y_pred, y_test, color = "green")
plt.title("Prediction vs Actual")
plt.xlabel("Predicted value")
plt.ylabel("Actual Value")
plt.show()
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test, y_pred)
print(cm)
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
#Naive Bayes model gives 91% accuracy on this dataset. We can further check accuracy using SVM model
```

### Prediction using SVM

```python
from sklearn.svm import SVC # "Support Vector Classifier" 
sv_obj = SVC(kernel='linear') 

# Model fit and predict
sv_model = sv_obj.fit(X_train,y_train)
y_pred2 = sv_model.predict(X_test)
```

```python
accuracy2=accuracy_score(y_test, y_pred2)
print("%.2f is the Accuracy obtained using SVM " %accuracy2)
```

#### Conclusion - We see that Naive Bayes performs better compared to SVM in this case giving heigher accuracy of 91%
